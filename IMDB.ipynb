# IMDB least budget


## IMDB.sqlite database contains all details of movies and has three tables with name IMDB, genre and earning.
## Find the least budgeted movie in IMDB Movie Database.
## If there is more than one least budgeted movies then print the movie which has maximum net profit.
## Note: Net Profit= Total Earning(Domestic+WorldWide earning) - budget

```python
data = pd.read_sql_query("SELECT Title, Budget, Domestic, Worldwide FROM IMDB INNER JOIN earning ON IMDB.Movie_id = earning.Movie_id", db)
data
```

```python
answer = data.values
answer
```

```python
for i in range(len(answer)):
    if (answer[i][0] == ''):
        answer[i][0] = '0'
    if (answer[i][1] == ''):
        answer[i][1] = '0'
    if (answer[i][2] == ''):
        answer[i][2] = '0'
answer
```

```python
netProfit = []
for i in range(len(answer)):
    netProfit.append(float(answer[i][2]) + float(answer[i][3]) - float(answer[i][1]))
budget = []
for i in range(len(answer)):
    budget.append(float(answer[i][1]))
movie_name = []
for i in range(len(answer)):
    movie_name.append(answer[i][0])
for i in range(len(answer)):
    print(netProfit[i], budget[i], movie_name[i])
```

```python
np_profit = np.array(netProfit)
np_budget = np.array(budget)
np_movie = np.array(movie_name)
```

```python
np_profit = np_profit[np.argsort(np_budget)]
np_movie = np_movie[np.argsort(np_budget)]
np_budget = np.sort(np_budget)
for i in range(len(np_budget)):
    print(np_budget[i], np_profit[i], np_movie[i])
```

# IMDB Year Top


## IMDB.sqlite' database contains all details of movies and has three tables with name IMDB, genre and earning.
## Find the highest-grossing movie (i.e domestic earning + worldwide earning) in IMDB database year wise.
## Print the Year and name of the movie which was the highest grosser in that year.
## Note:Print the year in ascending order.

```python
data = pd.read_sql_query("SELECT Title, Domestic, Worldwide FROM IMDB INNER JOIN earning ON IMDB.Movie_id = earning.Movie_id", db)
data
```

```python
answer = data.values
answer
```

```python
for i in range(len(answer)):
    if (answer[i][1] == ''):
        answer[i][1] = '0'
    if (answer[i][2] == ''):
        answer[i][2] = '0'
answer
```

```python
grossing = []
for i in range(len(answer)):
    grossing.append(float(answer[i][1]) + float(answer[i][2]))
year = []
for i in range(len(answer)):
    titleLength = len(answer[i][0])
    year.append(answer[i][0][titleLength - 5: titleLength - 1])
title = []
for i in range(len(answer)):
    title.append(answer[i][0])
```

```python
np_title = np.array(title)
np_grossing = np.array(grossing)
np_year = np.array(year)
```

```python
dictionary = {}
yearTitle = {}
for i in range(len(np_year)):
    if (np_year[i] in dictionary):
        if (np_grossing[i] > dictionary[np_year[i]]):
            yearTitle[np_year[i]] = np_title[i]
            dictionary[np_year[i]] = np_grossing[i]
    else:
        yearTitle[np_year[i]] = np_title[i]
        dictionary[np_year[i]] = np_grossing[i]
yearTitle
```

```python
import collections
```

```python
dictionary = collections.OrderedDict(sorted(yearTitle.items()))
dictionary
```

```python
for key, value in dictionary.items():
    print(key, value)
```

# IMDB genre budget


## IMDB.sqlite' database contains all details of movies and has three tables with name IMDB, genre and earning.
## Find out the percentage of the budget for each genre in IMDB Movie Dataset?Plot the pie chart.
## Print the genre and percentage of movies budget present in that genre with 2 decimal place after rounding off..
## Note: Movies has multiple genres, so calculate the percentage for each genre.Replace empty budget column to zero.

```python
data = pd.read_sql_query('SELECT genre, Budget FROM IMDB INNER JOIN genre ON IMDB.Movie_id = genre.Movie_id', db)
data
```

```python
answer = data.values
answer
```

```python
for i in range(len(answer)):
    if (answer[i][1] == ''):
        answer[i][1] = '0'
answer
```

```python
genre = []
for i in range(len(answer)):
    genre.append(answer[i][0])
genre
```

```python
budget = []
for i in range(len(answer)):
    budget.append(float(answer[i][1]))
budget
```

```python
dictionary = {}
for i in range(len(genre)):
    if (genre[i] == ''):
        continue
    if (genre[i] in dictionary):
        dictionary[genre[i]] += budget[i]
    else:
        dictionary[genre[i]] = budget[i]
dictionary
```

```python
genre = []
budget = []
for i in dictionary.keys():
    genre.append(i)
    budget.append(dictionary[i])
print(genre, budget)
```

```python
np_genre = np.array(genre)
np_budget = np.array(budget)
genre, budget
```

```python
plt.pie(budget, labels = genre, autopct = '%.2f')
plt.show()
```
